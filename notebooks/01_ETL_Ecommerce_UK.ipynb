```python
import pandas as pd          
import matplotlib.pyplot as plt  
import seaborn as sns       
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('../data/Raw/online_retail_II.csv')  
df.head()
```

```python

print(f"Filas: {df.shape[0]:,}")      
print(f"Columnas: {df.shape[1]}")      
print()


print("--- Tipos de datos ---")
print(df.dtypes)
print()

print("--- Valores faltantes por columna ---")
print(df.isnull().sum())
```

```python
filas_antes = len(df)


df = df.dropna(subset=['Customer ID'])


df = df[df['Quantity'] > 0]


df = df[df['Price'] > 0]


filas_despues = len(df)
print(f"Filas originales:  {filas_antes:,}")
print(f"Filas después de limpiar: {filas_despues:,}")
print(f"Filas eliminadas: {filas_antes - filas_despues:,} ({(filas_antes - filas_despues) / filas_antes * 100:.1f}%)")
```

```python

df['TotalVenta'] = df['Quantity'] * df['Price']

df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

df['Mes'] = df['InvoiceDate'].dt.month        
df['Anio'] = df['InvoiceDate'].dt.year         
df['DiaSemana'] = df['InvoiceDate'].dt.day_name()
traduccion_dias = {
    'Monday':    'Lunes',
    'Tuesday':   'Martes',
    'Wednesday': 'Miércoles',
    'Thursday':  'Jueves',
    'Friday':    'Viernes',
    'Saturday':  'Sábado',
    'Sunday':    'Domingo'
}
df['DiaSemana'] = df['DiaSemana'].replace(traduccion_dias)

df[['InvoiceDate', 'DiaSemana', 'Quantity', 'Price', 'TotalVenta', 'Mes', 'Anio']].head()
```
